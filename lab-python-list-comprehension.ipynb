````python
def initialize_inventory(products):
    # products is expected to be a list of strings (e.g., ['t-shirts', 'hats', 'keychain'])
    inventory = {
        product: int(input(f"Enter the quantity of {product} available: ")) 
        for product in products
    }
    return inventory
````

````python
def get_customer_orders():
    try:
        # Prompt for the number of orders
        num_orders = int(input("Enter the total number of customer orders: "))
    except ValueError:
        print("Invalid input. Assuming 0 orders.")
        return []

    # Use a List Comprehension to gather the product name for each order
    customer_orders = [
        input(f"Enter the name of product for order #{i+1}: ") 
        for i in range(num_orders)
    ]
    return customer_orders

# Example of how it should work (based on the image output):
# Enter the total number of customer orders: 4
# Enter the name of product for order #1: t-shirt
# Enter the name of product for order #2: hat
# Enter the name of product for order #3: t-shirt
# Enter the name of product for order #4: keychain
````

````python
def calculate_total_price(customer_orders):
    # Use a Generator Expression within the sum() function
    total_price = sum(
        float(input(f"Enter the price of {product}: "))
        for product in customer_orders
    )
    return total_price
````

````python
def update_inventory(inventory, customer_orders):
    # 1. Fulfill the orders (standard loop is best for this update)
    for product in customer_orders:
        if product in inventory and inventory[product] > 0:
            inventory[product] -= 1
        else:
            print(f"Warning: Cannot fulfill order for {product}. Item not in stock or not found.")

    # 2. Use Dictionary Comprehension to filter out products with zero quantity
    #    This creates a *new* dictionary that excludes items where the value is 0.
    updated_inventory = {
        product: quantity
        for product, quantity in inventory.items()
        if quantity > 0
    }
    
    return updated_inventory
````

````python
# --- Setup Data (Based on image input) ---
PRODUCTS = ['t-shirts', 'mug', 'hats', 'book', 'keychain']

# 1. Initialize Inventory
print("--- Inventory Initialization ---")
# User input for quantities: t-shirts: 5, hats: 3, keychain: 2
initial_inventory = initialize_inventory(PRODUCTS)

# 2. Get Customer Orders
print("\n--- Customer Orders ---")
# User input for 4 orders: t-shirt, hat, t-shirt, keychain
customer_orders = get_customer_orders()

# 3. Calculate Total Price
print("\n--- Price Calculation ---")
# User input for prices: t-shirt: 10.0, hat: 15.0, t-shirt: 10.0, keychain: 5.0
total_price = calculate_total_price(customer_orders)

# 4. Update Inventory
updated_inventory = update_inventory(initial_inventory, customer_orders)

# --- Output the Results ---
print("\n--- Order Statistics ---")
# Use a Set Comprehension to find unique products
unique_products_ordered = {product for product in customer_orders}
percentage_unique = (len(unique_products_ordered) / len(customer_orders)) * 100 if customer_orders else 0

print(f"Percentage of Unique Products Ordered: {percentage_unique:.1f}%")

print("\n--- Updated Inventory ---")
for product, quantity in updated_inventory.items():
    print(f"{product}: {quantity}")
    
print(f"\nTotal Price: {total_price:.2f}")

# Expected Output based on the image's example inputs:
# Initial Quantities: t-shirts: 5, hats: 3, keychain: 2
# Orders: t-shirt, hat, t-shirt, keychain (4 total orders)
# Inventory After Orders:
# t-shirts: 5 - 2 = 3
# hats: 3 - 1 = 2
# keychain: 2 - 1 = 1
# Prices: 10 + 15 + 10 + 5 = 40.0
# Unique Orders: t-shirt, hat, keychain (3 unique)
# Percentage: (3 / 4) * 100 = 75.0%
# (NOTE: The image output shows different numbers, perhaps a slight difference in initial input or the example is older.)
````
